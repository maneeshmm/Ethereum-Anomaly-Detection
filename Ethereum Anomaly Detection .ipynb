```python
print(df[missing_mask].shape)
df_sub = df[missing_mask]
```

```python
df_sub
```

```python
x =df_sub.drop('FLAG', axis = 1)
y = df_sub['FLAG']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state = 123)
```

```python
from category_encoders import TargetEncoder
```

```python
encoder = TargetEncoder(cols = categories)
```

```python
x_train_encoded = encoder.fit_transform(x_train,y_train)
```

```python
x_test_encoded = encoder.transform(x_test) 
```

```python
train = pd.concat([x_train_encoded, y_train], axis= 1)
```

```python
test = pd.concat([x_test_encoded, y_test], axis = 1)
```

```python
from sklearn.ensemble import IsolationForest
```

```python
isolation_forest = IsolationForest(contamination = 'auto', random_state = 123)
```

```python
isolation_forest.fit(train)
```

```python
anomaly_train_pred = isolation_forest.predict(train)
```

```python
anomaly_test_pred = isolation_forest.predict(test)
```

```python
is_train = np.sum(anomaly_train_pred == -1)
is_test = np.sum(anomaly_test_pred == 1)
```

```python
print('The number of outliers in training set:', is_train)
```

```python
print('The number of outliers in testing set:', is_test)
```

```python
from sklearn.manifold import TSNE
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
x_train_filled = imputer.fit_transform(x_train_encoded)
```

```python
tsne = TSNE(n_components=2, random_state=123)
x_train_tsne = tsne.fit_transform(x_train_filled)
```

```python
plt.scatter(x_train_tsne[:, 0], x_train_tsne[:, 1], alpha=0.5, c = anomaly_train_pred)
plt.title('t-SNE Visualization')
plt.show()
```

```python
x_test_filled = imputer.fit_transform(x_test_encoded)
x_test_tsne = tsne.fit_transform(x_test_filled)
plt.scatter(x_test_tsne[:, 0], x_test_tsne[:, 1], alpha=0.5, c=anomaly_test_pred)
plt.title('t-SNE Visualization (Test)')
```
